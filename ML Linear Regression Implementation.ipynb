```python
#prepare a Linear regression model to predict the GDP growth of Canadian Government using the 
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

```python
df=pd.read_csv('canada.csv')
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
sns.regplot(x='year',y='per capita income (US$)',data=df)
```

```python
x=df.drop('per capita income (US$)',axis=1)
```

```python
x
```

```python
y=df.drop('year',axis=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
brain=LinearRegression()
```
